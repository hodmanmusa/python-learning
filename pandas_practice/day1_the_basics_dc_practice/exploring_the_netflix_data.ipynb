## 2. Are movies more recent than TV Shows? 

```python
# Calcuating the mean of release date and comparing the result

movies_avg_release_year = movies_df["release_year"].describe()["mean"].item()
tv_avg_release_year = tv_shows["release_year"].describe()["mean"].item()

"Movies are more recent" if movies_avg_release_year>tv_avg_release_year else "Tv shows are more recent"
```

# üìò Section 4: Creating New Columns

## Tasks

1. Create:  
    1.1 content_age = 2026 - release_year  
    1.2 is_recent (True if release_year ‚â• 2018)  
    1.3 title_length  


```python
# Task 1.1 create column content_age = 2026 - release_year

netflix_df["content_age"] = 2026 - netflix_df["release_year"]
netflix_df[["title", "type", "release_year","content_age"]].head()
```

```python
# 1.2 is_recent (True if release_year ‚â• 2018)

netflix_df["is_recent"] = netflix_df["release_year"] >= 2018 
netflix_df[["title", "type", "release_year","is_recent"]].head(10)
```

```python
# Task 1.3 title_length

netflix_df["title_length"] = netflix_df["title"].str.len()
netflix_df[["title", "title_length"]].head()
```

# üìò Section 5: Basic Numerical Exploration (NumPy)

## Tasks
1. Use NumPy to compute:  
    1.1 Mean release year

```python
# Task 1.1 Mean release year

import numpy as np 

np.mean(netflix_df["release_year"]).item()
```

```python
# Task 1.2 Min & max release year

max_year = np.max(netflix_df["release_year"])
min_year = np.min(netflix_df["release_year"])

print("Min Year: ", min_year)
print("Max Year: ", max_year)
```

## Compare:
- Movies vs TV Shows average release year

```python
# Movies avrage release year 
movies_avg_yr = np.mean(movies_df["release_year"])

# TV Shows avrage release year 
tv_avg_yr = np.mean(tv_shows["release_year"])

# Compare
print("Movies Avrage Release Year: ", movies_avg_yr)
print("TV Show Avrage Release Year: ", tv_avg_yr)

print("Movie release year are more recent") if movies_avg_yr>tv_avg_yr else print("TV Show release year are more recent")
```

# üìò Section 6: Visualization Practice (Matplotlib)

## Charts to Create

1. Line chart: Number of titles released per year
2. Histogram: Distribution of release_year
3. Scatter plot: release_year vs content_age

```python
import matplotlib.pyplot as plt
```

```python
# Task 1 Line chart: Number of titles released per year
titles_per_year = netflix_df['release_year'].value_counts()
titles_per_year = titles_per_year.sort_index()

plt.plot(titles_per_year)
plt.title("Number of titles released per year")
plt.xlabel("Years")
plt.ylabel("Number of titles released")
plt.show()

# line_df["release_year"].value_counts()
```

```python
# Task 2: Histogram: Distribution of release_year
release_years = netflix_df["release_year"]

plt.hist(release_years, bins=20)
plt.title("Distribution of release year")
plt.xlabel("Years")
plt.ylabel("")
plt.show()
```

```python
# Task 3:  Scatter plot: release_year vs content_age

x = netflix_df['release_year']
y = netflix_df['content_age']

plt.scatter(x,y)
```

## Practice Questions
1. In which years did Netflix release more content?

```python
# Question 1: In which years did Netflix release more content

titles_per_year = netflix_df['release_year'].value_counts().sort_index()
titles_per_year.sort_values(ascending=False).head(5)
```

```python
# Question 2 Are older titles still common?
# Let's say "Older titles = released before 2000"

older_titles = netflix_df[netflix_df['release_year'] < 2000]
len(older_titles)
```

```python
len(older_titles) / len(netflix_df) * 100
```

<i>The porporation above states that only 5.96% of the titles are released before 2000. </i>
